```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv_imshow

img = cv.imread('/content/gdrive/MyDrive/MachineVision/lena.jpg')

# 원본영상에서 blur 제거
#계산한 값 uint8 범위 넘어갈 수도 있어서 int32로 계산
blur = cv.blur(img,ksize=(5,5))

# a=5
detail = np.int32(img) - np.int32(blur)

sharpened_img = np.int32(img) + 5*detail

cv_imshow(img)
cv_imshow(sharpened_img)
```

```python
# a=15
sharpened_img = np.int32(img) + 15*detail

cv_imshow(img)
cv_imshow(sharpened_img)
```

```python
# a=2
sharpened_img = np.int32(img) + 2*detail

cv_imshow(img)
cv_imshow(sharpened_img)
```
